```python
# Data Cleaning
import re
import pandas as pd


def clean_urdu(text):
    if pd.isna(text):
        return ""

    text = str(text)

    # Remove non-Urdu tweet noise before alignment.
    text = re.sub(r'http\S+|www\S+', ' ', text)
    text = re.sub(r'@\w+', ' ', text)
    text = re.sub(r'#\S+', ' ', text)
    text = re.sub(r'[A-Za-z0-9]', ' ', text)

    # Keep only Urdu letters. Turning punctuation into spaces avoids false pairs.
    text = re.sub(r'[^\u0600-\u06FF\s]', ' ', text)

    # Normalize spaces.
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df['CleanOriginal'] = df['TweetOriginal'].apply(clean_urdu)
df['CleanProcessed'] = df['TweetProcess'].apply(clean_urdu)

for i in range(10):
    print("\n====================")
    print("ORIGINAL CLEAN:")
    print(df.loc[i, 'CleanOriginal'])

    print("\nPROCESSED CLEAN:")
    print(df.loc[i, 'CleanProcessed'])
```

```python
import difflib


def find_all_changes(orig, proc):
    orig_words = str(orig).split()
    proc_words = str(proc).split()

    matcher = difflib.SequenceMatcher(
        a=orig_words,
        b=proc_words,
        autojunk=False
    )
    changes = []

    for tag, i1, i2, j1, j2 in matcher.get_opcodes():
        if tag != 'replace':
            continue

        wrong_block = orig_words[i1:i2]
        correct_block = proc_words[j1:j2]

        # Safe case: same number of words on both sides.
        if len(wrong_block) == len(correct_block):
            for wrong, correct in zip(wrong_block, correct_block):
                if wrong != correct:
                    changes.append((wrong, correct))

        # Safe case: exactly one original word replaced by exactly one processed word.
        elif len(wrong_block) == 1 and len(correct_block) == 1:
            wrong, correct = wrong_block[0], correct_block[0]
            if wrong != correct:
                changes.append((wrong, correct))

        # Unequal blocks are usually spacing/tokenization changes, so skip them.

    return changes
```

```python
df['Changes'] = df.apply(
    lambda row: find_all_changes(row['CleanOriginal'], row['CleanProcessed']),
    axis=1
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


value_counts = final_df['CorrectWord'].value_counts()
valid_words = value_counts[(value_counts >= 3) & (value_counts <= 1000)].index
final_df = final_df[final_df['CorrectWord'].isin(valid_words)]



final_df = final_df.copy()
final_df['Input'] = final_df['Sentence'] + " " + final_df['WrongWord']

final_df = final_df.dropna(subset=['Sentence', 'WrongWord', 'CorrectWord'])

final_df['Sentence'] = final_df['Sentence'].astype(str)
final_df['WrongWord'] = final_df['WrongWord'].astype(str)
final_df['CorrectWord'] = final_df['CorrectWord'].astype(str)

final_df['Input'] = final_df['Sentence'] + " " + final_df['WrongWord']

# 4. Vectorize
vectorizer = TfidfVectorizer(ngram_range=(1,3), max_features=10000)
X = vectorizer.fit_transform(final_df['Input'])
y = final_df['CorrectWord']

print("After filtering:", len(final_df))

```

```python
# ================================
# RNN MODEL
# ================================

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, top_k_accuracy_score

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.utils import to_categorical

# --------------------------------
# Prepare Data
# --------------------------------

X = final_df['Input'].astype(str)
y = final_df['CorrectWord'].astype(str)

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

num_classes = len(label_encoder.classes_)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    
)

# --------------------------------
# Tokenization
# --------------------------------

max_words = 10000
max_len = 20

tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')

# One-hot encode labels
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)

# --------------------------------
# Build RNN Model
# --------------------------------

rnn_model = Sequential([
    Embedding(input_dim=max_words, output_dim=64, input_length=max_len),

    SimpleRNN(64),

    Dense(64, activation='relu'),

    Dense(num_classes, activation='softmax')
])

rnn_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

rnn_model.summary()

# --------------------------------
# Train Model
# --------------------------------

history = rnn_model.fit(
    X_train_pad,
    y_train_cat,
    epochs=10,
    batch_size=32,
    validation_split=0.1
)

# --------------------------------
# Predictions
# --------------------------------

y_pred_probs = rnn_model.predict(X_test_pad)

y_pred = np.argmax(y_pred_probs, axis=1)

# --------------------------------
# Metrics
# --------------------------------

accuracy = accuracy_score(y_test, y_pred)

precision = precision_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0
)

recall = recall_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0
)

f1 = f1_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0
)


# --------------------------------
# Results
# --------------------------------

print("\n===== RNN RESULTS =====")

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")


print("\nClassification Report:\n")

print(
    classification_report(
        y_test,
        y_pred,
        zero_division=0
    )
)
```

```python
# ================================
# LSTM MODEL
# ================================

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    top_k_accuracy_score
)

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping

# --------------------------------
# REMOVE VERY RARE CLASSES
# --------------------------------

value_counts = final_df['CorrectWord'].value_counts()

valid_words = value_counts[value_counts >= 6].index

final_df = final_df[
    final_df['CorrectWord'].isin(valid_words)
]

# --------------------------------
# PREPARE DATA
# --------------------------------

X = final_df['Input'].astype(str)

y = final_df['CorrectWord'].astype(str)

# Encode labels
label_encoder = LabelEncoder()

y_encoded = label_encoder.fit_transform(y)

num_classes = len(label_encoder.classes_)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42
)

# --------------------------------
# TOKENIZATION
# --------------------------------

max_words = 15000

max_len = 30

tokenizer = Tokenizer(
    num_words=max_words,
    oov_token="<OOV>"
)

tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)

X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(
    X_train_seq,
    maxlen=max_len,
    padding='post'
)

X_test_pad = pad_sequences(
    X_test_seq,
    maxlen=max_len,
    padding='post'
)

# --------------------------------
# ONE HOT LABELS
# --------------------------------

y_train_cat = to_categorical(
    y_train,
    num_classes=num_classes
)

y_test_cat = to_categorical(
    y_test,
    num_classes=num_classes
)

# --------------------------------
# BUILD LSTM MODEL
# --------------------------------

lstm_model = Sequential([

    Embedding(
        input_dim=max_words,
        output_dim=128,
        input_length=max_len
    ),

    LSTM(
        128,
        dropout=0.3,
        recurrent_dropout=0.3
    ),

    Dense(128, activation='relu'),

    Dropout(0.3),

    Dense(num_classes, activation='softmax')
])

# --------------------------------
# COMPILE
# --------------------------------

lstm_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

lstm_model.summary()

# --------------------------------
# EARLY STOPPING
# --------------------------------

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

# --------------------------------
# TRAIN
# --------------------------------

history = lstm_model.fit(
    X_train_pad,
    y_train_cat,
    epochs=20,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stop]
)

# --------------------------------
# PREDICTIONS
# --------------------------------

y_pred_probs = lstm_model.predict(X_test_pad)

y_pred = np.argmax(y_pred_probs, axis=1)

# --------------------------------
# METRICS
# --------------------------------

accuracy = accuracy_score(y_test, y_pred)

precision = precision_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0
)

recall = recall_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0
)

f1 = f1_score(
    y_test,
    y_pred,
    average='weighted',
    zero_division=0
)

top3_acc = top_k_accuracy_score(
    y_test,
    y_pred_probs,
    k=3,
    labels=np.arange(num_classes)
)

# --------------------------------
# RESULTS
# --------------------------------

print("\n===== LSTM RESULTS =====")

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")
print(f"Top-3 Acc : {top3_acc:.4f}")

print("\nClassification Report:\n")

print(
    classification_report(
        y_test,
        y_pred,
        zero_division=0
    )
)
```
